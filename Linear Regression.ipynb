```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
cali=datasets.fetch_california_housing()
df=pd.DataFrame(cali.data,columns=cali.feature_names)
df['price']=cali.target
df.head()
```

```python
input = df["MedInc"]
output = df["price"]
```

```python
plt.scatter(input,output)
```

```python
len(df["price"])
```

```python
input = np.array(df["AveRooms"])
output = np.array(df["price"])
input.shape
```

```python
input = input.reshape(-1,1)
output = output.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_output, test_output=train_test_split(input,output)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
```

```python
LR.fit(train_input,train_output)
```

```python
LR.coef_, LR.intercept_
```

```python
plt.scatter(input,output)
plt.plot(input,LR.coef_*input + LR.intercept_, "r")
```
