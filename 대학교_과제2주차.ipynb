```python

"""#2. 深層NN

プログラム 2.1

(a) Irisデータセットの70%を訓練、30%をテストとする。
"""

import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd

# データ読み込み
iris = load_iris()
X = iris.data
y = iris.target

# 訓練とテストに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# 区切り数をいくつか試す
bin_list = [2, 3, 4, 5, 6, 7, 8, 9, 10]
results = []

for bins in bin_list:
    # 各特徴量の最小値と最大値に基づいてビンの境界を設定
    bin_edges = [np.linspace(np.min(X[:, i]), np.max(X[:, i]), bins + 1) for i in range(4)]

    # 訓練データからセルごとのラベル出現回数を記録
    cell_table = {}

    for xi, yi in zip(X_train, y_train):
        # 各特徴量をビンに割り当てる
        cell = tuple(np.digitize(xi[i], bin_edges[i]) - 1 for i in range(4))
        if cell not in cell_table:
            cell_table[cell] = np.zeros(3)  # クラスは3つ
        cell_table[cell][yi] += 1

    # テストデータを分類
    correct = 0
    unknown = 0

    for xi, yi in zip(X_test, y_test):
        cell = tuple(np.digitize(xi[i], bin_edges[i]) - 1 for i in range(4))
        if cell not in cell_table or np.sum(cell_table[cell]) == 0:
            unknown += 1  # 訓練データに現れなかったセル
        else:
            pred = np.argmax(cell_table[cell])
            if pred == yi:
                correct += 1

    accuracy = correct / len(X_test)
    unknown_ratio = unknown / len(X_test)
    results.append((bins, accuracy, unknown_ratio))

# 結果を表にして表示
df_results = pd.DataFrame(results, columns=["1特徴あたりのビン数", "精度", "未知セル率"])
print(df_results)

"""プログラム 2.2

sklearn.neural_network を用いて3値分類の学習を行う。2層、3層、4層の学習を行い、その精度を比較する。

(a) Irisデータセットの70%を訓練、30%をテストとする。

(b) model = MLPClassifier(hidden_layer_sizes=XXX, max_iter=1000, random_state=0)のXXXをそれぞれ(), (5, ), (5, 4)とする。
"""

from sklearn.neural_network import MLPClassifier

# 2層NN: 隠れ層なし（入力→出力 直接）
model = MLPClassifier(hidden_layer_sizes=(), max_iter=1000, random_state=0)
model.fit(X_train, y_train)

# 精度確認
acc = model.score(X_test, y_test)
print(f"Test Accuracy: {acc:.3f}")

# 重みの取得
for i, coef in enumerate(model.coefs_):
    print(f"\n重みW{i+1} (layer {i} → {i+1}):\n", coef)

for i, bias in enumerate(model.intercepts_):
    print(f"\nバイアスb{i+1} (layer {i+1}):\n", bias)

# 3層NN: 入力-隠れ層(5ノード)-出力（softmax）で分類
model = MLPClassifier(hidden_layer_sizes=(5,), max_iter=1000, random_state=0)
model.fit(X_train, y_train)

# 4層NN: 入力 → 隠れ層(5) → 隠れ層(4) → 出力
model = MLPClassifier(hidden_layer_sizes=(5, 4), max_iter=1000, random_state=0)
model.fit(X_train, y_train)

"""プログラム 2.3

3層NN（隠れ層が1層、3ノード、活性化関数はtanh）に$(-1,1)$を$N$等分した入力とその関数の値(雑音は加えない)の対を$N$組与えて関数を学習させる。そのときの中間層の出力を表示して、中間層が出力層にどのような貢献をしているかを明示する。

(a) 赤線：ニューラルネットワークの出力（近似関数）

(b) 青い点：訓練データ（$N=50$ 個）

(c) 破線：3つの隠れユニットの出力（tanh活性）

$f(x) = x^2$（非線形な滑らかな曲線）

$f(x) = \sin(x)$（周期関数）
"""

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

# Define the target functions
functions = {
    "x^2": lambda x: x**2,
    "sin(x)": lambda x: np.sin(x)
}

# Prepare input data
N = 50
x = np.linspace(-1, 1, N).reshape(-1, 1)

# Set up the plot
fig, axes = plt.subplots(1, 2, figsize=(12, 8))
axes = axes.ravel()

# Iterate over each function
for idx, (label, func) in enumerate(functions.items()):
    y = func(x.ravel())

    # Create and train the MLPRegressor
    model = MLPRegressor(hidden_layer_sizes=(3,), activation='tanh', solver='lbfgs', max_iter=10000, random_state=0)
    model.fit(x, y)

    # Generate predictions for plotting
    x_plot = np.linspace(-1, 1, 400).reshape(-1, 1)
    y_pred = model.predict(x_plot)

    # Hidden unit activations (before output layer)
    hidden_activation = np.tanh(np.dot(x_plot, model.coefs_[0]) + model.intercepts_[0])

    # Plot results
    ax = axes[idx]
    ax.plot(x_plot, y_pred, 'r', label='ネットワーク出力')
    ax.plot(x, y, 'bo', label='訓練データ')
    for i in range(3):
        ax.plot(x_plot, hidden_activation[:, i], '--', label=f'隠れ層 {i+1}')
    ax.set_title(f"{label}の近似")
    ax.legend(loc='best')
    ax.grid(True)

plt.tight_layout()
plt.show()

"""プログラム 2.4

手書き数字画像 (digits データセット) を使って、あるデータ点（アンカー）と同じクラスの特徴（正例）を近づけ、異なるクラスの特徴（負例）からは遠ざけるような表現（埋め込み）を学習する。学習する前と後で、InfoNCEの値にどのような変化が生じたかを確認する。

(a) データを準備する

手書き数字データ（digits）を用意

特徴量（64次元）を標準化（平均0、分散1）

(b)生データでInfoNCE損失を計算

まだ何も学習していない状態

生データをそのまま使って、InfoNCE損失を計算する

ここでの損失は大きい → クラスの分離が悪い

(c) 表現学習のためのニューラルネットを作る

小さなニューラルネット

入力 → 隠れ層（ReLU）→ 埋め込み層（64次元）

出力ベクトルが「学習された表現（特徴）」になる

(d) 表現学習を始める

ミニバッチ学習

目的は「InfoNCE損失を小さくすること」

(d) 学習後の表現でInfoNCE損失を計算

学習したニューラルネットから出力を得る

その出力ベクトルを使ってInfoNCE損失を再計算
"""

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

# --- データ準備 ---
digits = load_digits()
X = StandardScaler().fit_transform(digits.data)  # (nサンプル × 64次元) に標準化
y = digits.target  # ラベル（0〜9）

# --- コサイン類似度行列を計算（ライブラリを使わずに） ---
def cosine_similarity_matrix(X):
    X_norm = np.linalg.norm(X, axis=1, keepdims=True)  # 各行のノルムを計算
    X_normalized = X / (X_norm + 1e-8)  # 正規化（ゼロ割防止に微小値を加える）
    return X_normalized @ X_normalized.T  # 類似度行列 (n×n)

# --- InfoNCE損失を計算 ---
def info_nce(z, labels):
    sim = cosine_similarity_matrix(z)
    N = len(z)
    loss = 0
    for i in range(N):
        pos_mask = (labels == labels[i])
        pos_mask[i] = False  # 自分自身は除く
        numerator = np.sum(np.exp(sim[i, pos_mask]))
        denominator = np.sum(np.exp(sim[i])) - np.exp(sim[i, i])
        if numerator > 0 and denominator > 0:
            loss += -np.log(numerator / denominator)
    return loss / N

# --- 最終隠れ層出力を抽出（特徴ベクトル） ---
def extract_features(model, X):
    h = X
    for i in range(len(model.coefs_) - 1):  # 隠れ層までReLU適用
        h = np.maximum(0, h @ model.coefs_[i] + model.intercepts_[i])
    return h @ model.coefs_[-1] + model.intercepts_[-1]  # 最後は線形

# --- 学習前のInfoNCE損失（生特徴） ---
loss_before = info_nce(X, y)

# --- シンプルなMLP（自己再構成目的） ---
mlp = MLPRegressor(hidden_layer_sizes=(64,), activation='relu', max_iter=100, random_state=0)
mlp.fit(X, X)  # オートエンコーダ風に学習
Z = extract_features(mlp, X)  # 学習後の特徴ベクトル
loss_after = info_nce(Z, y)

# --- 結果表示 ---
print(f"学習前のInfoNCE損失: {loss_before:.4f}")
print(f"学習後のInfoNCE損失: {loss_after:.4f}")

"""プログラム 2.5

混合分布ネットワークの動作を確認する

(a) $y=x+0.3\sin(2\pi x) + ([-0.1,-.1]の一様分布)$によって生成した$N=50$サンプルからNNで回帰を行う。

(b) しかし、$(x,y)$を入れ替えて$(y,x)$とする(逆問題)と、各$x$に対して複数通り(最大3個)の$y$が存在する(多峰性)ので、推定が難しい。

(c) そこで、混合正規分布の推定問題に帰着して、各正規分布の$(\pi_k, \mu_k, \sigma^2_k$),  $\pi_k: {\mathbb R}^M\rightarrow [0,1]$, $\mu_k: {\mathbb R}^M\rightarrow {\mathbb R}^L$,
$\sigma^2_k: {\mathbb R}^M\rightarrow {\mathbb R}_{>0}$, $k=1,2,3$を学習する。

最初にサンプルを生成する。
"""

import numpy as np
import matplotlib.pyplot as plt

# --- 1. データ生成（順問題：x → y） ---
np.random.seed(0)  # 再現性のための乱数シード
N = 50  # サンプル数

# x を [0,1] の一様分布からサンプリング
x = np.random.uniform(0, 1, N)

# 正解関数：y = x + 0.3 * sin(2πx) + ノイズ（[-0.1, 0.1]）
y = x + 0.3 * np.sin(2 * np.pi * x) + np.random.uniform(-0.1, 0.1, N)

# --- 2. 学習データを並べ替え（プロットを滑らかにするため） ---
idx = np.argsort(x)
x_sorted = x[idx]
y_sorted = y[idx]

"""(a)は、例えば、以下のようにして実現できる。"""

from sklearn.neural_network import MLPRegressor

# --- 3. 多層パーセプトロン（MLP）モデルの定義 ---
model = MLPRegressor(hidden_layer_sizes=(50, 50),  # 隠れ層を2層に
                     activation='tanh',            # 活性化関数 tanh
                     solver='lbfgs',               # 最適化アルゴリズム
                     max_iter=10000,               # 学習の繰り返し回数
                     random_state=0)

# --- 4. 学習（入力x、出力y） ---
model.fit(x.reshape(-1, 1), y)

# --- 5. 予測と可視化 ---
x_plot = np.linspace(0, 1, 200).reshape(-1, 1)  # テスト入力点（滑らかにプロットするため）
y_pred = model.predict(x_plot)  # 予測値

# --- 6. 結果の表示 ---
plt.figure(figsize=(8, 4))
plt.scatter(x, y, color='blue', label='訓練データ', alpha=0.7)
plt.plot(x_plot, y_pred, color='red', linewidth=2, label='MLPによる予測')
plt.title("順問題の回帰： $x \\rightarrow y$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

"""(b)は問題11で検討する。(c)は、下記で実現できる。GaussianMixtureには要素数=3を指定する。"""

from sklearn.mixture import GaussianMixture

# --- 2. 入出力を反転： y → x を学習する逆問題へ ---
X_input = y.reshape(-1, 1)  # 入力は y（横軸）
y_target = x                # 出力は x（縦軸）

# --- 3. GMMによる条件付き分布の近似 ---
# GMMは (y, x) の2次元分布に対するモデル
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=0)
gmm.fit(np.hstack([X_input, y_target.reshape(-1, 1)]))  # 入力(y), 出力(x)

# --- 4. 可視化用グリッドの生成 ---
yg, xg = np.meshgrid(
    np.linspace(min(y)-0.1, max(y)+0.1, 100),
    np.linspace(min(x)-0.1, max(x)+0.1, 100)
)
grid = np.c_[yg.ravel(), xg.ravel()]
logprob = gmm.score_samples(grid)
pdf = np.exp(logprob).reshape(yg.shape)

# --- 5. 結果の表示 ---
plt.figure(figsize=(8, 5))
plt.contourf(yg, xg, pdf, levels=20, cmap='Blues')
plt.scatter(y, x, color='black', s=20, alpha=0.6, label="訓練データ")
plt.title("MDN風の多峰性分布近似： $y \\rightarrow x$")
plt.xlabel("y")
plt.ylabel("x")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

# 9번: f(x) = |x| 및 f(x) = H(x) (Heaviside 함수) 학습 및 시각화

# 입력 데이터 준비
N = 50
x = np.linspace(-1, 1, N).reshape(-1, 1)

# 학습할 함수 사전 정의
functions = {
    "|x|": lambda x: np.abs(x),
    "H(x)": lambda x: np.where(x >= 0, 1.0, 0.0)
}

# 2개의 서브플롯 생성
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes = axes.ravel()

for idx, (label, func) in enumerate(functions.items()):
    # 타깃 값 생성
    y = func(x.ravel())

    # MLPRegressor 모델 생성
    model = MLPRegressor(
        hidden_layer_sizes=(3,),  # 은닉층 3유닛
        activation='tanh',        # 활성화 함수 tanh
        solver='lbfgs',           # 최적화 알고리즘
        max_iter=10000,
        random_state=0
    )
    model.fit(x, y)

    # 예측용 촘촘한 x
    x_plot = np.linspace(-1, 1, 400).reshape(-1, 1)
    y_pred = model.predict(x_plot)

    # 은닉 유닛 활성화 (tanh)
    hidden_act = np.tanh(x_plot @ model.coefs_[0] + model.intercepts_[0])

    # 그래프 그리기
    ax = axes[idx]
    ax.plot(x_plot, y_pred, 'r-', label='NN Output')
    ax.scatter(x, y, c='b', label='Training Data')
    for i in range(3):
        ax.plot(x_plot, hidden_act[:, i], '--', label=f'Hidden Unit {i+1}')
    ax.set_title(f"{label} Approximation")
    ax.set_xlabel("x")
    ax.set_ylabel("y / Activation")
    ax.legend(loc='best')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# --- データ準備 ---
digits = load_digits()
X = StandardScaler().fit_transform(digits.data)  # 平均0・分散1に標準化
y = digits.target

# --- InfoNCE損失の計算関数 ---
def cosine_similarity_matrix(Z):
    Z_norm = np.linalg.norm(Z, axis=1, keepdims=True)
    Z_normalized = Z / (Z_norm + 1e-8)
    return Z_normalized @ Z_normalized.T

def info_nce(z, labels):
    sim = cosine_similarity_matrix(z)
    N = len(z)
    loss = 0.0
    for i in range(N):
        pos = (labels == labels[i])
        pos[i] = False
        num = np.sum(np.exp(sim[i, pos]))
        den = np.sum(np.exp(sim[i])) - np.exp(sim[i, i])
        loss += -np.log(num / den)
    return loss / N

# --- オートエンコーダ風 MLP で実験 ---
hidden_dims = [64, 128, 256]
results = []

for dim in hidden_dims:
    # --- 日本語コメント: オートエンコーダ学習 ---
    mlp = MLPRegressor(hidden_layer_sizes=(dim,),
                       activation='relu',
                       solver='lbfgs',
                       max_iter=200,
                       random_state=0)
    mlp.fit(X, X)

    # --- 日本語コメント: 隠れ層出力の抽出 ---
    h = X
    for w, b in zip(mlp.coefs_[:-1], mlp.intercepts_[:-1]):
        h = np.maximum(0, h @ w + b)
    Z = h @ mlp.coefs_[-1] + mlp.intercepts_[-1]

    loss_val = info_nce(Z, y)
    results.append((dim, loss_val))

# --- 結果表示 ---
df = pd.DataFrame(results, columns=["hidden_dim", "InfoNCE_loss"])
print(df)
```

隠蔽層ユニット数を64、128、256に増やして実験した結果、InfoNCE損失はそれぞれ約1.9436→1.9438→1.9437で、64で最も低く、128で少し上昇した後、256で再び少し減少する傾向を示しました。これは、隠蔽次元が小さいと表現力が不足してアンダーフィッティングが発生し、クラス間の区分が適切に行われないためであり、次元が増加すると、入力の再構成性能は上がりますが、InfoNCEが要求する「類似サンプルは近くに、異質サンプルは遠くに」する埋め込み分離力は一部低下し、損失が増加する可能性があることを意味します。 また、次元が大きすぎると、ノイズやサンプル固有の詳細情報まで学習して過適合を引き起こし、embeddingの一般化能力を低下させます。


```python
# --- データ生成はそのまま ---
np.random.seed(0)
N = 50
x = np.random.uniform(0, 1, N)
y = x + 0.3 * np.sin(2 * np.pi * x) + np.random.uniform(-0.1, 0.1, N)

# --- (2) 入出力を逆問題に変更 ---
# Japanese コメント: 逆問題では入力に y，出力に x を使う
X_input = y.reshape(-1, 1)  # 入力は y
y_target = x                # 出力は x

# --- (3) モデル構築は共通 ---
model = MLPRegressor(hidden_layer_sizes=(50, 50),
                     activation='tanh',
                     solver='lbfgs',
                     max_iter=10000,
                     random_state=0)

# --- (4) 学習 ---
model.fit(X_input, y_target)

# --- (5) 可視化の入力範囲設定 ---
# Japanese コメント: y の最小値～最大値でプロット用点を作成
y_plot = np.linspace(np.min(y), np.max(y), 200).reshape(-1, 1)

# --- 予測 ---
x_pred = model.predict(y_plot)

# --- (6) プロット ---
plt.figure(figsize=(8, 5))
# Japanese コメント: 散布図は (y, x)
plt.scatter(y, x, color='blue', label='訓練データ', alpha=0.7)
# Japanese コメント: 予測線は (y_plot, x_pred)
plt.plot(y_plot, x_pred, color='red', linewidth=2, label='MLP逆問題予測')
plt.title("逆問題の回帰： $y \\rightarrow x$")
plt.xlabel("y")
plt.ylabel("x")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
